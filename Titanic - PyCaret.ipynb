```python
def age(df):
    df.loc[ df['Age'] <= 11, 'Age'] = 0
    df.loc[(df['Age'] > 11) & (df['Age'] <= 18), 'Age'] = 1
    df.loc[(df['Age'] > 18) & (df['Age'] <= 22), 'Age'] = 2
    df.loc[(df['Age'] > 22) & (df['Age'] <= 27), 'Age'] = 3
    df.loc[(df['Age'] > 27) & (df['Age'] <= 33), 'Age'] = 4
    df.loc[(df['Age'] > 33) & (df['Age'] <= 40), 'Age'] = 5
    df.loc[(df['Age'] > 40) & (df['Age'] <= 66), 'Age'] = 6
    df.loc[ df['Age'] > 66, 'Age'] = 6
```

```python
age(train)
```

```python
def dp(df):
    df.drop(['Name'],axis=1,inplace=True)
    df.drop(['Ticket'],axis=1,inplace=True)
    df.drop(['Cabin'],axis=1,inplace=True)
    df.drop(['Fare'],axis=1,inplace=True)
```

```python
dp(train)
```

```python
dp(test)
```

```python
from pycaret import classification
```

```python
set_up = classification.setup(data=train,target='Survived')
```

```python
# Build decision tree model
dtree = classification.create_model('dt')
```

```python
xgb = classification.create_model('xgboost')
```

```python
cat = classification.create_model('catboost')
```

```python
tune_dtree = classification.tune_model(dtree)
```

```python
tune_xgb= classification.tune_model(xgb)
```

```python
tune_cat = classification.tune_model(cat)
```

```python
# Emsemble Dtree
boosting_dtree = classification.ensemble_model(dtree,method='Boosting')
```

```python
boosting_xgb = classification.ensemble_model(xgb,method='Boosting')
```

```python
boosting_cat = classification.ensemble_model(cat,method='Boosting')
```

```python
blending = classification.blend_models(estimator_list=[dtree,xgb,cat])
```

## Compare Models

```python
classification.compare_models()
```

```python
# evaluate model - all Evaluation method
classification.evaluate_model(cat)
```

```python
# interpret: SHAP
classification.interpret_model(cat)
```

```python
# interpret: correlation
classification.interpret_model(cat,plot='correlation')
```
